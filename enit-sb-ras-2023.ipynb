```python
df = pd.read_csv('/kaggle/input/road-traffic-accidents/RTA Dataset.csv')
df['Time'] = pd.to_datetime(df['Time'])
df.head()
# Plot the histogram
df['Time'].dt.hour.plot(kind='hist', bins=24, rwidth=0.8, color='pink')
plt.title('Distribution of Times')
plt.xlabel('Hour of the Day')
plt.ylabel('Frequency')
plt.show()
```

```python
hour_frequency = df['Time'].dt.hour.value_counts().sort_index()
hour_frequency=hour_frequency/sum(hour_frequency)
print("Hour Frequency:")
print(hour_frequency)
hour_frequency.plot(kind='bar', color='violet')
plt.title('Hourly Frequency Distribution')
plt.xlabel('Hour of the Day')
plt.ylabel('Frequency')
plt.show()
```

```python
days_frequency = df['Day_of_week'].value_counts().sort_index()
days_frequency=days_frequency/sum(days_frequency)
print("days_frequency:")
print(days_frequency)
days_frequency.plot(kind='bar', color='orange')
plt.title('Daily Frequency Distribution')
plt.xlabel('Day of the week')
plt.ylabel('Frequency')
plt.show()
```
